```python
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
```

```python
plt.figure(figsize=(15,80), facecolor='white')
plotnumber=1
for categorical_feature in categorical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(y= categorical_feature, data=dep)
    plt.xlabel(categorical_feature)
    plt.title(categorical_feature)
    plotnumber+=1
plt.show( )    
```

```python
for categorical_feature in categorical_features:
    sns.catplot(x= 'subscribed', col=categorical_feature, kind= 'count', data=dep)
plt.show( )
```

```python
for categorical_feature in categorical_features:
    print(dep.groupby(['subscribed', categorical_feature]).size( ))
```

```python
numerical_features = [feature for feature in dep.columns if ((dep[feature].dtypes != 'object') & (feature not in ['subscribed']))]
print('Number of numerical variables: ', len(numerical_features))

dep[numerical_features].head( )
```
